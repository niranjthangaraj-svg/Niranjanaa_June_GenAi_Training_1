```python
!pip install -q google-generativeai

import google.generativeai as genai
from google.colab import userdata

genai.configure(api_key=userdata.get('Generative_Ai'))

knowledge_base = """
travel_agency: WanderWave Travels

=== Bookings ===
{
    "question": "How do I book a trip?",
    "answer": "You can book a trip by visiting our website www.wanderwavetravels.com, selecting your destination, choosing travel dates, and completing the payment."
},
{
    "question": "Can I book offline?",
    "answer": "Yes, you can visit our nearest branch office or call our helpline to book your trip."
},
{
    "question": "Can I modify my booking?",
    "answer": "Yes, bookings can be modified up to 48 hours before departure, subject to availability and charges."
},
{
    "question": "How do I cancel my trip?",
    "answer": "You can cancel through your booking dashboard or contact support. Cancellation charges may apply based on timing."
},

=== Packages ===
{
    "question": "What types of travel packages are available?",
    "answer": "We offer domestic tours, international tours, honeymoon packages, adventure tours, and pilgrimage packages."
},
{
    "question": "Do you offer custom packages?",
    "answer": "Yes, we provide fully customized travel itineraries based on your budget and preferences."
},

=== Payments ===
{
    "question": "What payment methods are accepted?",
    "answer": "We accept UPI, credit/debit cards, net banking, and EMI options for selected packages."
},
{
    "question": "Is advance payment required?",
    "answer": "Yes, a minimum of 30% advance is required to confirm a booking."
},

=== Visa & Documents ===
{
    "question": "Do you help with visa processing?",
    "answer": "Yes, we assist with tourist visa applications for most international destinations."
},
{
    "question": "What documents are required?",
    "answer": "Valid passport, ID proof, passport-size photos, and visa-specific documents depending on the destination."
},

=== Hotels & Transport ===
{
    "question": "Are hotels included in packages?",
    "answer": "Yes, most packages include hotel accommodation based on the selected category (budget, standard, luxury)."
},
{
    "question": "Do you provide flight bookings?",
    "answer": "Yes, we provide complete flight booking assistance with competitive fares."
},
{
    "question": "Is local transport included?",
    "answer": "Yes, selected packages include airport transfers and local sightseeing transport."
},

=== Contact ===
Contact: support@wanderwavetravels.com | Helpline: 1800-987-6543 (Mon-Sun, 8 AM - 8 PM)
Bookings: bookings@wanderwavetravels.com
Visa Support: visa@wanderwavetravels.com
"""

system_instruction = f"""
You are a friendly and helpful travel agency chatbot for WanderWave Travels.

Follow these rules strictly:

1. Answer ONLY using the information in the knowledge base below.
2. If a question is not covered in the knowledge base, reply exactly:
   "I'm sorry, I don't have that information. Please contact support@wanderwavetravels.com or call 1800-987-6543."
3. Be polite, concise, and helpful.
4. Do not make up any information outside the knowledge base.

{knowledge_base}
"""

model = genai.GenerativeModel(
    model_name="gemini-2.5-flash-lite",
    system_instruction=system_instruction
)

chat = model.start_chat(history=[])

print("WanderWave Travel Bot is ready! Type 'quit' to exit.\n")

while True:
    user_input = input("You: ").strip()

    if not user_input:
        continue

    if user_input.lower() in ["quit", "exit", "bye"]:
        print("Bot: Thank you for choosing WanderWave Travels! Safe journeys")
        break

    response = chat.send_message(user_input)
    print(f"Bot: {response.text}\n")
```
